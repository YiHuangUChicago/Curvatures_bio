```python
nums = np.array([5, 10, 15, 20, 25, 30])

cs=7

plt.figure(figsize=(10, 7))
plt.tick_params(axis='both', which='major', labelsize=20)

y = [bm[n][0] for n in nums]
yerr = [bm[n][1] for n in nums]
plt.errorbar(nums, y, yerr=yerr, fmt="o-", capsize=cs, label='AdaBoost')

r = 431 / 434.0
y = [record_1[n][0] * r  for n in nums]
yerr = [record_1[n][1] * r for n in nums]
plt.errorbar(nums-.5, y, yerr=yerr, fmt="o-", capsize=cs, label='1-dimensional Gaussian')

y = [record_2[n][0] * r  for n in nums]
yerr = [record_2[n][1] * r for n in nums]
plt.errorbar(nums+.5, y, yerr=yerr, fmt="o-", capsize=cs, label='2-dimensional Gaussian')

plt.legend(prop={'size': 20})
plt.savefig('Congress_labelProp_vs_AdaBoost', transparent=True ,bbox_inches='tight')
plt.show()
```
